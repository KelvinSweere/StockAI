```python
s_tick = input("Ticker name from Yahoo Finance... ")
#RDSA.AS --> Shell NL
# s_tick = "RDSA.AS"
```

```python
import yfinance as yf

tckr = yf.Ticker(s_tick)
print(tckr)
```

```python
try:
    tckr_info = tckr.get_info()
    
except KeyError:
    print("Ticker " + str(s_tick) + " is found.")
except IndexError:
    print("No information found about ticker " + str(s_tick))
    
else:
    print(tckr_info['sector'].lower() + " - ticker has been founded.")
```

```python
#get data from a month ago...
from datetime import date
import dateutil.relativedelta
today_date = date.today()

last_month_date = (today_date + dateutil.relativedelta.relativedelta(months=-1))
```

```python
from pandas_datareader import data as pdr
from pandas.util.testing import assert_frame_equal

yf.pdr_override() # <== that's all it takes :-)

# download dataframe
data = pdr.get_data_yahoo(s_tick, start=last_month_date, end=today_date)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.plot(data['Open'],'ro-', label='line 1', linewidth=0.5)
plt.plot(data['Close'],'bo-', label='line 1', linewidth=0.5)
plt.grid(True)
plt.xlabel("Date")
plt.ylabel("â‚¬")
plt.show()
```

```python
data.insert(6, "Delta", data['Open']-data['Close'], True) 
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data['Delta'],'go-', label='line 1', linewidth=0.5)
plt.grid(True)
plt.xlabel("Date")
plt.ylabel("â‚¬")
plt.show()
```

```python
import numpy as np
X_day_num = []
for row in data.index: 
    X_day_num.append(row.strftime("%d")) 

np.array(X_day_num).reshape(-1, 1)
print(X_day_num)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit([X_day_num],data['Open'])
```
