```python
# s_tick = input("Ticker name from Yahoo Finance... ")
#RDSA.AS --> Shell NL
s_tick = "RDSA.AS"
```

```python
import yfinance as yf

tckr = yf.Ticker(s_tick.upper())
print(tckr)
```

```python
try:
    tckr_info = tckr.get_info()
    
except KeyError:
    print("Ticker " + str(s_tick) + " is found.")
except IndexError:
    print("No information found about ticker " + str(s_tick))
    
else:
    print(tckr_info['sector'].lower() + " - ticker has been founded.")
```

```python
#get data from a month ago...
from datetime import date
import dateutil.relativedelta
today_date = date.today()

last_month_date = (today_date + dateutil.relativedelta.relativedelta(months=-1))
```

```python
from pandas.util.testing import assert_frame_equal

data = tckr.history(period="2d", interval='1m', action=False)
```

```python
#feature engineering
#rendament

def rendament(new_value, old_value):
    return round(((new_value-old_value)/old_value)*100,2)

data.insert(7, "Day rendament", rendament(data['Close'],data['Open']), True)
```
